## 决策树实战task:
- 基于iris_data.csv数据，建立决策树模型，评估模型表现
- 可视化决策树结构
- 修改min_samples_leaf参数，对比模型结果

```python
# load the data 
import pandas as pd
import numpy as np
data = pd.read_csv('iris_data.csv')
data.head()
```

```python
# define the X and y
X = data.drop(['target','label'],axis = 1)
y = data.loc[:,'label']
print(X.shape,y.shape)
```

```python
# establish the decision tree model
from sklearn import tree
dc_tree = tree.DecisionTreeClassifier(criterion = 'entropy',min_samples_leaf = 5)
dc_tree.fit(X,y)
```

```python
# evaluate the model
y_predict = dc_tree.predict(X)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
print(accuracy)
```

```python
# visualize the tree
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize = (10,10))
tree.plot_tree(dc_tree,filled = 'True',feature_names = ['SepalLength','SepalWidth','PetalLength','PetalWidth'],class_names = ['setosa','versicolor','virginica'])
```

决策树实战summary:
- 通过建立决策树，可实现对标签数据的有效分类
- 通过修改叶子节点最少样本对应参数min_samples_leaf,可调控制树的分支深度；
